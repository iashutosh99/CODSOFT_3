# Under Sampling
Builed a sample dataset containing similar distribution of normal transaction and fraudlend transaction

Number of Fraudlent Transaction --> 492

```python
legit_sample = legit.sample(n = 492)
```

# Concatinating two DataFrame

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

# Selecting the data into Feature and Target

```python
X = new_data.drop(columns="Class",axis = 1)
y = new_data["Class"]
```

```python
X
```

```python
y
```

# Spliting the data in training data and test data

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```


# Model training

```python
lr = LogisticRegression()
svm = SVC(probability=True)
RF = RandomForestClassifier()
DT = DecisionTreeClassifier()
NB = GaussianNB()
ensemble= VotingClassifier(estimators=[
    ('lr', lr),
    ('svm', svm),
    ('rf',RF ),
     ('dt',DT ),
     ('nb', NB )

], voting='soft')
```

```python
#Training the LogisticRegression model with training data
lr.fit(X_train,y_train)
svm.fit(X_train,y_train)
RF.fit(X_train,y_train)
DT.fit(X_train,y_train)
NB.fit(X_train,y_train)
ensemble.fit(X_train,y_train)
```


#Model Evaluation
Accuracy score

```python
X_train_prediction = lr.predict(X_train)
lr_accuracy = accuracy_score(y_train,X_train_prediction)
lr_precision= precision_score(y_train,X_train_prediction)
lr_recall= recall_score(y_train,X_train_prediction)
print("Accuracy score on training data by LR: ",lr_accuracy)
print("Precision score on training data by LR: ",lr_precision)
print("recall score score on training data by LR: ",lr_recall)
print()

X_train_prediction = svm.predict(X_train)
accuracy = accuracy_score(y_train,X_train_prediction)
precision= precision_score(y_train,X_train_prediction)
recall= recall_score(y_train,X_train_prediction)
print("Accuracy score on training data by SVM: ",accuracy)
print("Precision score on training data by SVM: ",precision)
print("recall score score on training data by SVM: ",recall)
print()

X_train_prediction = RF.predict(X_train)
accuracy = accuracy_score(y_train,X_train_prediction)
precision= precision_score(y_train,X_train_prediction)
recall= recall_score(y_train,X_train_prediction)
print("Accuracy score on training data by RF: ",accuracy)
print("Precision score on training data by RF: ",precision)
print("recall score score on training data by RF: ",recall)
print()

X_train_prediction = DT.predict(X_train)
accuracy = accuracy_score(y_train,X_train_prediction)
precision= precision_score(y_train,X_train_prediction)
recall= recall_score(y_train,X_train_prediction)
print("Accuracy score on training data by DT: ",accuracy)
print("Precision score on training data by DT: ",precision)
print("recall score score on training data by DT: ",recall)
print()

X_train_prediction = NB.predict(X_train)
accuracy = accuracy_score(y_train,X_train_prediction)
precision= precision_score(y_train,X_train_prediction)
recall= recall_score(y_train,X_train_prediction)
print("Accuracy score on training data by NB: ",accuracy)
print("Precision score on training data by NB: ",precision)
print("recall score score on training data by NB: ",recall)
print()

ensemble_predictions = ensemble.predict(X_train)
accuracy = accuracy_score(y_train, ensemble_predictions)
precision_score = precision_score(y_train, ensemble_predictions)
recall_score = recall_score(y_train, ensemble_predictions)

print("Ensemble Classifier Accuracy:", accuracy)
print("Ensemble Classifier precision:", precision)
print("Ensemble Classifier recall:", recall)
```
